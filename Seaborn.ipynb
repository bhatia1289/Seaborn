```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
sns.get_dataset_names()
```

```python
#how to read dataset
df = sns.load_dataset("tips")
```

```python
# numerical data plotting
#relplot has options of plotting both categorical and numerical data
# replot >> how the relationship between two variables
df
```

```python
#how tip is varying wrt total bill > scatterplot / by default replot makes scatter plot
sns.relplot(x='total_bill', y = 'tip', data = df) #scatterplot
```

```python
sns.relplot(x='total_bill', y = 'tip', data = df, hue = 'smoker')
#customer who gave the highest tip is a smoker
#hue gives different colorset
```

```python
sns.relplot(x='total_bill', y = 'tip', data = df, hue='size')
```

```python
sns.relplot(x='total_bill', y = 'tip', data = df, style= 'size')
#size gives different represntation of a column value
```

```python
#use hue and style both at same time
```

```python
sns.relplot(x='total_bill', y = 'tip', data = df, hue='time', style='smoker')
```

```python
#relplot for making a line plot >> generating a data
#line plot >> to understand the trend in data , a time column /sequence is generally involved
```

```python
time = np.arange(500)
value = np.random.randn(500).cumsum()
df = pd.DataFrame({'time' : time, 'value' : value})
```

```python
sns.relplot(x='time', y='value', kind = 'line', data = df , sort = True)
```

```python
#joints plot>> understand relationship and also distribution of two variables
df = sns.load_dataset("tips")
sns.jointplot(x= df.total_bill , y = df.tip, kind = 'reg') #reg>> regression line to represent relationship
```

```python
sns.jointplot(x= df.total_bill , y = df.tip, kind = 'kde') #kernal density estimator , contour plot >> metrological department
```
